```python
# Read in the CSV files
val_maes_dir = "./result_transfer/lr1e-5/model_A_B_val_maes_batch.csv"
train_maes_dir = "./result_transfer/lr1e-5/model_A_B_maes_batch.csv"

import csv
with open(val_maes_dir, 'r') as f:
    df1 = list(csv.reader(f, delimiter=";"))
with open(train_maes_dir, 'r') as f:
    df2 = list(csv.reader(f, delimiter=";"))

import numpy as np
df1 = np.array(df1)
df2 = np.array(df2)

data_val = []
for i in range(len(df2)):
    value = float(df1[i][0][1:-1])
    data_val.append(value)
    
data_train = []
for i in range(len(df2)):
    value = float(df2[i][0][1:-1])
    data_train.append(value)   

# Create the x-axis

iteration1 = range(len(data_val))
iteration2 = range(len(data_train))

# Plot the data
fig = plt.figure(figsize=(10,4))
plt.plot(data_val, label='Validation MAE')
plt.plot(data_train,label='Training MAE')

# Add axis labels and a legend
plt.xlabel('Iteration')
plt.ylabel('MAE')
plt.title("MAE per Batch for Model A trasnfer Learning with B")
plt.legend()

# Show the plot
plt.show()
if save:
    fig.savefig('./result_transfer/plots/MAE_Batch_ModelA_B.png')
```
